<font color=red> <font size=4> ** Sift features extraction ** </font>

```python
# define function which allow to convert the image to grey scale.
def to_gray(img):
    """This function transform the introduced image into grey scale image.
    
       Args:
       
       img(Image):The introduced image , which the function will transform.
       
       Returns:
       
       gray(Image): The photo transformed into a grey scale photo.
    """
    gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
    return gray
```

```python
#Create features extraction and keypoint detector object.
def gen_sift_features(gray_img):
    """This function could extract features and keypoints for each image.
    
       Args:
       
       gray_img: The introduced gray imgage, which the function will extract its features and keypoints.
       
       Returns:
       
       kp,desc: the tuple returned by this function, correponding respectively to the keypoint and the features 
                descriptors of the introduced image.
    """
    sift=cv.xfeatures2d.SIFT_create()  # features extractor.
    kp,desc=sift.detectAndCompute(gray_img,None)
    return kp,desc
```

```python
#Create a function which allow to show the features of each image.
def show_sift_features(grey_img,color_img,kp):
    """This function allow to highlight features in the image.
    
    Args:
    
    grey_img(array) :transformed image to grey picture.
    color_img(Image): the origin image.
    kp(array): the keypoint features of the image.
    
    Returns:
    
    im(Image): the grey image with highlighting of the keypoints features of the image.
    """
    return plt.imshow(cv.drawKeypoints(grey_img,kp,color_img.copy()))
```

```python
#Hereunder,we will make preview in what they could make the previous function in our images. Thus,we will see 
#the grey picture of the above choiced image, we will see also their keyfeatures extracted by the SIFT alghorithm.
img=images[10]

grey_img=to_gray(img) 
kp,desc=gen_sift_features(grey_img)
fig=plt.figure(figsize=(15,30))
fig.add_subplot(1,3,1)
plt.imshow(img)
plt.title('The origin image',size=14,color='blue')
fig.add_subplot(1,3,2)
plt.imshow(grey_img)
plt.title('Grey picture image',size=14,color='blue')
fig.add_subplot(1,3,3)
show_sift_features(grey_img,img,kp)
plt.title('Image SIFT features',size=14,color='blue')
plt.show()
```

```python
# We will now see how the SIFT features match up each other across two images.
img=images[10]
other_img=images[20]
grey_other=to_gray(other_img)
kp_other,des_other=gen_sift_features(grey_other)
bf=cv.BFMatcher(cv.NORM_L2,crossCheck=True)
matches=bf.match(desc,des_other)
matches=sorted(matches,key=lambda x:x.distance) # sort the matches in the order of their distances.
N_Matches=100 # draw the top N matches.
match_img=cv.drawMatches(img,kp,other_img,kp_other,matches[:N_Matches],img.copy(),flags=0)
plt.figure(figsize=(12,6))
plt.imshow(match_img)
```

<font size =4> <font color='grey'> **We will now apply all the obove operations to the whole images, in order to establish the bag of visual words of each one ** </font>.

```python
des_list=[]   #list where all the descriptors are stored.
for i, im in enumerate(images):
    im_g=to_gray(im)
    kp,descr=gen_sift_features(im_g) # computing the keypoints and the descriptors.
    des_list.append((i,descr)) # appending all the descriptors into the single list.
```

```python
# Stack all the descriptors vertically in a numpy array.
descriptors=des_list[0][1]
for elt in des_list[1:]:
    descriptors=np.vstack((descriptors,elt[1]))
```

```python
#Perform k-means clustering.
from sklearn.cluster import KMeans
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import silhouette_score ,make_scorer
# Determinate the optimum number of cluster to use on the base of the silhouette score.
#params=[100,300,500,700,900,1000]
model=KMeans(n_clusters=500)
model.fit(descriptors)
  
```

<font color=blue><font size=4> **Unfortunatly, we couldn't compute the performance of this clustering because of the limite memory of our machine**</font>

```python
#calculate the histogram of features.
bag_features=np.zeros((len(images),500))
for elt in des_list:
    words=model.predict(elt[1])
    for w in words:
        bag_features[elt[0]][w]+=1
```

<font color=red> <font size=3> ** Classification **</font>

```python
#Split the dataset on training dataset and test dataset.
from sklearn.model_selection import train_test_split
X=bag_features
Y=np.array(labels)
xtr,xts,ytr,yts=train_test_split(X,Y,train_size=0.6,stratify=Y)
```

```python
#We will use to classify our dogs, the method SVM.
#We will determinate, first of all the best hyperparametrs of this function.
from sklearn.svm import SVC
params={'C':np.logspace(-3,3)}
mod=SVC()
lr=GridSearchCV(mod,params,cv=5,scoring='accuracy')
lr.fit(xtr,ytr)
print ('The best penality parameter to use is: {}'.format(lr.best_params_))
```

```python
#Compute the performance of this approach.
performance=lr.score(xts,yts)
print('The performance of classical approach is :{}'.format(performance))
```
